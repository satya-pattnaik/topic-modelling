```
import pandas as pd

df = pd.read_csv("voted-kaggle-dataset.csv")
df.head()
```

```

import spacy
import gensim
```

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```
df['lemmatized_sentence'] = df['Description'].apply(lambda sentence:gensim.utils.simple_preprocess(str(sentence),deacc=False))
df.head()

```

```
# Remove stopwords

df['stopwords_removed'] = df['lemmatized_sentence'].apply(lambda listOfWords: [word for word in listOfWords if word not in stopwords])
df.head()[df.columns[-3:]]
```

```
bigramPhraser = gensim.models.Phrases(df['lemmatized_sentence'].tolist(), min_count=5, threshold=100)
```

```
#Create Dictionary of all words 
id2word = gensim.corpora.Dictionary(df['stopwords_removed'].tolist()) #id to word mapping
df['bag_of_words'] = df['stopwords_removed'].apply(lambda words:id2word.doc2bow(words)) #Maps id of word to its freq in the doc.
df.head()[df.columns[-4:]]
```

```
df['count_frequency'] = df['bag_of_words'].apply(lambda words:[(id2word[word[0]],word[1]) for word in words]) #Maps word to its freq in the doc.
df.head()[df.columns[-5:]]
```

```
# Build the topic model
corpus = df['bag_of_words'].tolist()
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=20, #The number of requested latent topics to be extracted from the training corpus.
                                           random_state=100,
                                           update_every=1,#Number of documents to be iterated through for each update. Set to 0 for batch learning, > 1 for online iterative learning
                                           chunksize=100,#Number of documents to be used in each training chunk
                                           passes=100,
                                           alpha='auto',#’auto’: Learns an asymmetric prior from the corpus.
                                           per_word_topics=True)
```

```
from pprint import pprint
pprint(lda_model.print_topics())
```

```
lda_model.log_perplexity(corpus)
```
